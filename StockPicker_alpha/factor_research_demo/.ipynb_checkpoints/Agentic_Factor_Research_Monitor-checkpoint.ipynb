# Agentic Factor Research Monitor: From Hypothesis to Backtest Review

This notebook is a compact factor research flight recorder. It does not only ask whether a factor backtest looks good. It asks whether the research process itself is trustworthy enough to keep investigating.

The workflow builds three monthly equity factor strategies, measures their backtest behavior, and runs a structured research monitor that flags leakage risk, overfitting symptoms, regime instability, weak out-of-sample behavior, and next validation steps.

## 1. Research Setup

The notebook uses deterministic synthetic data so the review experience is reproducible and does not depend on network access. The synthetic universe has market regimes, stock-level factor exposures, noisy returns, and a lagged valuation proxy.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 140)

RANDOM_SEED = 42
rng = np.random.default_rng(RANDOM_SEED)

TICKERS = [
    'AAPL', 'MSFT', 'NVDA', 'AMZN', 'META', 'GOOGL', 'JPM', 'V', 'MA', 'UNH',
    'HD', 'PG', 'KO', 'PEP', 'COST', 'WMT', 'XOM', 'CVX', 'CAT', 'BA',
    'GE', 'IBM', 'INTC', 'CSCO', 'ADBE', 'CRM', 'NFLX', 'DIS', 'MCD', 'NKE'
]
START = '2017-01-03'
END = '2026-05-29'
BENCHMARK_NAME = 'Synthetic Market Benchmark'

print(f'Universe size: {len(TICKERS)} stocks')
print(f'Sample: {START} to {END}')
print(f'Benchmark: {BENCHMARK_NAME}')
```

```python
def regime_label(date):
    if date < pd.Timestamp('2020-01-01'):
        return 'pre_2020'
    if date < pd.Timestamp('2022-01-01'):
        return 'covid_rebound'
    if date < pd.Timestamp('2025-01-01'):
        return 'inflation_tightening'
    return 'recent'

def generate_synthetic_equity_panel(tickers, start, end, seed=42):
    rng = np.random.default_rng(seed)
    dates = pd.bdate_range(start, end)
    n = len(tickers)

    static = pd.DataFrame({
        'ticker': tickers,
        'quality_exposure': rng.normal(0, 1, n),
        'value_exposure': rng.normal(0, 1, n),
        'risk_exposure': rng.uniform(0.75, 1.45, n),
    }).set_index('ticker')

    regime_params = {
        'pre_2020': {'market_mu': 0.00035, 'market_vol': 0.0080, 'value': 0.00003, 'quality': 0.00004, 'risk_penalty': -0.00002},
        'covid_rebound': {'market_mu': 0.00055, 'market_vol': 0.0140, 'value': -0.00006, 'quality': 0.00006, 'risk_penalty': 0.00001},
        'inflation_tightening': {'market_mu': 0.00010, 'market_vol': 0.0110, 'value': 0.00007, 'quality': 0.00002, 'risk_penalty': -0.00005},
        'recent': {'market_mu': 0.00024, 'market_vol': 0.0095, 'value': 0.00002, 'quality': 0.00003, 'risk_penalty': -0.00001},
    }

    returns = pd.DataFrame(index=dates, columns=tickers, dtype=float)
    benchmark = pd.Series(index=dates, dtype=float, name='benchmark')

    for date in dates:
        rp = regime_params[regime_label(date)]
        market = rng.normal(rp['market_mu'], rp['market_vol'])
        benchmark.loc[date] = market
        idio = rng.normal(0, 0.009, n)
        cross_sectional_alpha = (
            rp['value'] * static['value_exposure'].to_numpy()
            + rp['quality'] * static['quality_exposure'].to_numpy()
            + rp['risk_penalty'] * static['risk_exposure'].to_numpy()
        )
        returns.loc[date] = market * static['risk_exposure'].to_numpy() + cross_sectional_alpha + idio

    prices = 100 * (1 + returns).cumprod()
    benchmark_prices = 100 * (1 + benchmark).cumprod()

    month_ends = prices.resample('M').last().index
    valuation_noise = pd.DataFrame(
        rng.normal(0, 0.35, (len(month_ends), n)), index=month_ends, columns=tickers
    )
    value_score = pd.DataFrame(
        static['value_exposure'].to_numpy()[None, :] + valuation_noise.to_numpy(),
        index=month_ends,
        columns=tickers,
    )

    monthly_features = {
        'value_score': value_score.shift(1),
        'static_exposures': static,
    }
    return prices, returns, benchmark, benchmark_prices, monthly_features

prices, daily_returns, benchmark_returns, benchmark_prices, monthly_features = generate_synthetic_equity_panel(TICKERS, START, END, RANDOM_SEED)

display(prices.tail())
benchmark_prices.plot(figsize=(10, 3), title='Synthetic Market Benchmark')
plt.ylabel('Index Level')
plt.show()
```

## 2. Factor Construction

The strategies rebalance monthly and use only information available before the holding period. Each factor forms a dollar-neutral long-short portfolio using the top and bottom quintiles of the cross-section.

```python
monthly_prices = prices.resample('M').last()
monthly_returns = monthly_prices.pct_change()

factor_scores = {
    'Momentum_12_1': monthly_prices.pct_change(12).shift(1),
    'Low_Volatility': -daily_returns.rolling(126).std().resample('M').last().shift(1),
    'Value_Proxy': monthly_features['value_score'],
}

factor_hypotheses = {
    'Momentum_12_1': 'Stocks with strong trailing 12-month returns, excluding the latest month, may continue to outperform due to slow information diffusion and investor underreaction.',
    'Low_Volatility': 'Lower-risk stocks may earn better risk-adjusted returns because leverage constraints and benchmark incentives can push investors toward high-beta names.',
    'Value_Proxy': 'Stocks with cheaper valuation characteristics may outperform as prices mean-revert toward fundamentals, though the effect can be regime-dependent.',
}

def build_long_short_portfolio(scores, next_returns, q=0.2):
    rows = []
    holdings = {}
    for date in scores.index.intersection(next_returns.index):
        s = scores.loc[date].dropna()
        r = next_returns.loc[date].dropna()
        common = s.index.intersection(r.index)
        s = s.loc[common]
        r = r.loc[common]
        if len(common) < 10:
            continue
        n_bucket = max(3, int(np.floor(len(common) * q)))
        ranked = s.sort_values()
        short_names = ranked.head(n_bucket).index
        long_names = ranked.tail(n_bucket).index
        weight = pd.Series(0.0, index=common)
        weight.loc[long_names] = 1 / n_bucket
        weight.loc[short_names] = -1 / n_bucket
        rows.append({'date': date, 'return': float((weight * r).sum()), 'longs': tuple(long_names), 'shorts': tuple(short_names)})
        holdings[date] = set(long_names).union(set(short_names))
    portfolio = pd.DataFrame(rows).set_index('date')
    portfolio['gross_exposure'] = 2.0
    return portfolio, holdings

next_month_returns = monthly_returns.shift(-1)
factor_portfolios = {}
factor_holdings = {}
for name, scores in factor_scores.items():
    pf, holdings = build_long_short_portfolio(scores, next_month_returns)
    factor_portfolios[name] = pf
    factor_holdings[name] = holdings

summary = pd.DataFrame({name: pf['return'] for name, pf in factor_portfolios.items()})
display(summary.tail())
```

## 3. Backtest Metrics

The core metrics are designed to support a PM-style conversation: performance, risk, path dependency, hit rate, turnover, and whether performance survives different regimes.

```python
def max_drawdown(returns):
    equity = (1 + returns.dropna()).cumprod()
    drawdown = equity / equity.cummax() - 1
    return float(drawdown.min())

def annualized_metrics(returns):
    r = returns.dropna()
    if r.empty:
        return {'cagr': np.nan, 'sharpe': np.nan, 'max_drawdown': np.nan, 'hit_rate': np.nan}
    years = len(r) / 12
    total_return = (1 + r).prod() - 1
    cagr = (1 + total_return) ** (1 / years) - 1 if years > 0 else np.nan
    sharpe = np.sqrt(12) * r.mean() / r.std() if r.std() > 0 else np.nan
    return {
        'cagr': cagr,
        'sharpe': sharpe,
        'max_drawdown': max_drawdown(r),
        'hit_rate': float((r > 0).mean()),
    }

def calculate_turnover(holdings):
    dates = sorted(holdings.keys())
    values = []
    for prev, curr in zip(dates[:-1], dates[1:]):
        previous = holdings[prev]
        current = holdings[curr]
        if not previous and not current:
            values.append(0.0)
        else:
            values.append(1 - len(previous.intersection(current)) / len(previous.union(current)))
    return float(np.mean(values)) if values else np.nan

metrics = []
for name, returns in summary.items():
    m = annualized_metrics(returns)
    m['avg_turnover'] = calculate_turnover(factor_holdings[name])
    m['factor'] = name
    metrics.append(m)

metrics_df = pd.DataFrame(metrics).set_index('factor')
display(metrics_df.style.format({'cagr': '{:.2%}', 'sharpe': '{:.2f}', 'max_drawdown': '{:.2%}', 'hit_rate': '{:.2%}', 'avg_turnover': '{:.2%}'}))

equity_curves = (1 + summary.dropna(how='all')).cumprod()
equity_curves.plot(figsize=(11, 4), title='Long-Short Factor Equity Curves')
plt.ylabel('Growth of $1')
plt.show()
```

```python
def month_regime(date):
    if date < pd.Timestamp('2020-01-01'):
        return 'pre_2020'
    if date < pd.Timestamp('2022-01-01'):
        return '2020_2021'
    if date < pd.Timestamp('2025-01-01'):
        return '2022_2024'
    return 'recent'

regime_rows = []
for factor in summary.columns:
    factor_returns = summary[factor].dropna()
    for regime, r in factor_returns.groupby(lambda d: month_regime(pd.Timestamp(d))):
        row = annualized_metrics(r)
        row['factor'] = factor
        row['regime'] = regime
        regime_rows.append(row)

regime_df = pd.DataFrame(regime_rows).set_index(['factor', 'regime']).sort_index()
display(regime_df.style.format({'cagr': '{:.2%}', 'sharpe': '{:.2f}', 'max_drawdown': '{:.2%}', 'hit_rate': '{:.2%}'}))

rolling_sharpe = summary.rolling(24).mean() / summary.rolling(24).std() * np.sqrt(12)
rolling_sharpe.plot(figsize=(11, 4), title='24-Month Rolling Sharpe')
plt.axhline(0, color='black', linewidth=1)
plt.ylabel('Sharpe')
plt.show()
```

## 4. Agentic Research Monitor

The monitor is intentionally transparent and rule-based. In a production agentic research system, these checks could be delegated to specialized agents: data integrity, methodology, performance diagnostics, regime analysis, and PM memo generation.

```python
def severity_rank(label):
    return {'Low': 1, 'Medium': 2, 'High': 3}.get(label, 0)

def monitor_factor(factor, returns, metrics_df, regime_df, rolling_sharpe):
    m = metrics_df.loc[factor]
    factor_regimes = regime_df.loc[factor]
    rs = rolling_sharpe[factor].dropna()
    rows = []

    rows.append({
        'risk_flag': 'Hypothesis clarity',
        'evidence': factor_hypotheses[factor],
        'severity': 'Low',
        'next_action': 'Keep the economic rationale attached to every backtest variant.'
    })

    rows.append({
        'risk_flag': 'Lookahead control',
        'evidence': 'Signals are shifted before the next-month holding period; value proxy is lagged by one month.',
        'severity': 'Low',
        'next_action': 'In live data, verify point-in-time fundamentals and corporate-action handling.'
    })

    if m['avg_turnover'] > 0.70:
        rows.append({
            'risk_flag': 'High turnover',
            'evidence': f'Average monthly holdings turnover is {m["avg_turnover"]:.1%}.',
            'severity': 'High',
            'next_action': 'Add transaction costs, rebalance buffers, and capacity assumptions.'
        })
    elif m['avg_turnover'] > 0.45:
        rows.append({
            'risk_flag': 'Moderate turnover',
            'evidence': f'Average monthly holdings turnover is {m["avg_turnover"]:.1%}.',
            'severity': 'Medium',
            'next_action': 'Estimate transaction cost sensitivity before treating the alpha as tradable.'
        })

    if m['sharpe'] < 0.30:
        rows.append({
            'risk_flag': 'Weak full-sample signal',
            'evidence': f'Full-sample Sharpe is only {m["sharpe"]:.2f}.',
            'severity': 'High',
            'next_action': 'Reject or redesign the signal unless a stronger sub-universe rationale exists.'
        })
    elif m['sharpe'] > 2.50:
        rows.append({
            'risk_flag': 'Suspiciously strong result',
            'evidence': f'Full-sample Sharpe is {m["sharpe"]:.2f}, which is unusually high for a simple public factor.',
            'severity': 'High',
            'next_action': 'Audit data generation, timestamp alignment, survivorship assumptions, and cost model.'
        })

    recent_sharpe = factor_regimes.loc['recent', 'sharpe'] if 'recent' in factor_regimes.index else np.nan
    earlier_sharpe = factor_regimes.drop(index='recent', errors='ignore')['sharpe'].mean()
    if pd.notna(recent_sharpe) and pd.notna(earlier_sharpe) and recent_sharpe < earlier_sharpe - 0.75:
        rows.append({
            'risk_flag': 'Weak recent out-of-sample behavior',
            'evidence': f'Recent Sharpe is {recent_sharpe:.2f} versus earlier average Sharpe of {earlier_sharpe:.2f}.',
            'severity': 'High',
            'next_action': 'Investigate crowding, regime dependence, and whether the signal needs conditioning.'
        })

    regime_sharpes = factor_regimes['sharpe'].dropna()
    if len(regime_sharpes) >= 3 and regime_sharpes.min() < -0.25 and regime_sharpes.max() > 0.75:
        rows.append({
            'risk_flag': 'Regime instability',
            'evidence': f'Regime Sharpe ranges from {regime_sharpes.min():.2f} to {regime_sharpes.max():.2f}.',
            'severity': 'Medium',
            'next_action': 'Test macro-conditioned allocation or combine with complementary factors.'
        })

    if len(rs) > 0 and (rs < 0).mean() > 0.35:
        rows.append({
            'risk_flag': 'Unstable rolling Sharpe',
            'evidence': f'{(rs < 0).mean():.1%} of 24-month rolling Sharpe observations are negative.',
            'severity': 'Medium',
            'next_action': 'Run walk-forward validation and compare against a simple equal-weight benchmark.'
        })

    if m['max_drawdown'] < -0.20:
        rows.append({
            'risk_flag': 'Material drawdown',
            'evidence': f'Max drawdown is {m["max_drawdown"]:.1%}.',
            'severity': 'Medium',
            'next_action': 'Review drawdown timing and whether risk controls would have reduced exposure.'
        })

    report = pd.DataFrame(rows)
    report['severity_rank'] = report['severity'].map(severity_rank)
    return report.sort_values(['severity_rank', 'risk_flag'], ascending=[False, True]).drop(columns='severity_rank')

monitor_reports = {
    factor: monitor_factor(factor, summary[factor], metrics_df, regime_df, rolling_sharpe)
    for factor in summary.columns
}

for factor, report in monitor_reports.items():
    print(f'\n=== {factor} Research Monitor ===')
    display(report)
```

## 5. Final Research Memo

The memo is generated from the computed metrics and monitor reports. This is the bridge from a technical notebook to an asset-management research conversation.

```python
def choose_best_factor(metrics_df, monitor_reports):
    score = metrics_df['sharpe'].fillna(-99) + 0.5 * metrics_df['hit_rate'].fillna(0) - metrics_df['avg_turnover'].fillna(1)
    for factor, report in monitor_reports.items():
        score.loc[factor] -= 0.35 * (report['severity'] == 'High').sum()
        score.loc[factor] -= 0.12 * (report['severity'] == 'Medium').sum()
    return score.sort_values(ascending=False).index[0], score.sort_values(ascending=False)

best_factor, research_scores = choose_best_factor(metrics_df, monitor_reports)
best_report = monitor_reports[best_factor]
top_risks = best_report[best_report['severity'].isin(['High', 'Medium'])]
if top_risks.empty:
    top_risk_text = 'The monitor did not find major blocking issues, but live-data validation is still required.'
else:
    top_risk_text = '; '.join(top_risks.head(3)['risk_flag'].tolist())

memo = f"""
# Investment Research Memo

## Recommended factor for continued research
**{best_factor}** is the strongest candidate after combining performance, hit rate, turnover, and monitor penalties.

## Why it might work
{factor_hypotheses[best_factor]}

## Evidence from the backtest
- CAGR: {metrics_df.loc[best_factor, 'cagr']:.2%}
- Sharpe: {metrics_df.loc[best_factor, 'sharpe']:.2f}
- Max drawdown: {metrics_df.loc[best_factor, 'max_drawdown']:.2%}
- Hit rate: {metrics_df.loc[best_factor, 'hit_rate']:.2%}
- Average monthly turnover: {metrics_df.loc[best_factor, 'avg_turnover']:.2%}

## What the monitor caught
{top_risk_text}

## What a PM or researcher should do next
1. Re-run the same workflow on point-in-time live data with survivorship-bias controls.
2. Add transaction costs and liquidity constraints before accepting the result as tradable.
3. Run walk-forward validation and compare against simpler baselines.
4. Test whether the factor should be combined with complementary factors rather than traded alone.

## Research governance takeaway
The value of the agentic layer is not that it magically finds alpha. Its value is that it makes the research process auditable: every attractive backtest is paired with explicit assumptions, risk flags, and next tests.
"""

print(memo)
display(pd.DataFrame({'research_score': research_scores}).style.format({'research_score': '{:.2f}'}))
```
